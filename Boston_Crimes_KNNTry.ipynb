```python
##getting seasons for visualization

def Seasons(mon):
    if (mon == 12 or mon == 1 or mon == 2):
       return "Winter"
    elif(mon == 3 or mon == 4 or mon == 5):
       return "Spring"
    elif(mon ==6 or mon==7 or mon == 8):
       return "Summer"
    else:
       return "Fall"
```

```python
df['Season'] =df.loc[:,'MONTH'].apply(Seasons).sort_values(ascending=False)
df.head()
```

```python
##visualizing in terms of seasons
season_counts = df.groupby('Season')['OFFENSE_CODE_GROUP'].count().sort_values(ascending=False).to_frame().reset_index()
ax = sns.barplot(x = 'Season' , y = "OFFENSE_CODE_GROUP",data = season_counts, palette='YlGnBu')
plt.title('Number Of Crimes Each Season');
```

```python
X = df.drop(['DISTRICT'] , axis=1)
Y = df.DISTRICT

x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.10,random_state = 100)

sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

```python
##Applying KNN with 10 neighbors 
knn = KNeighborsClassifier(n_neighbors=12, metric='minkowski')
knn.fit(X_train , y_train)
y_pred_knn = knn.predict(X_test)

cm_knn = confusion_matrix(y_test , y_pred_knn)
print('KNN')
print(cm_knn)
```
